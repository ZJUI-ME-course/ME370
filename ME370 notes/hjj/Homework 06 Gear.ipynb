$$\because 1 kph=\frac{1000m}{3600 s}$$  

$$\therefore 1kph \cdot \frac{3600}{1000}=1m/s$$  

$$\therefore 3.6 kph =1m/s $$

```python
from math import pi,sin,cos,tan,asin,acos,atan
kg=1
W=1
rpm=2*pi/60
mm=0.001
du=pi/180
g=9.81
kph=3.6

N_pedal=42
m=90*kg
angle=15*du
W_max=300*W
w_pedal=80*rpm
r_pedal=170*mm
d_wheel=700*mm

r_wheel=d_wheel/2 #注意上面给的是直径


F_uphill=m*g*sin(angle)
v_uphill=W_max/F_uphill

w_wheel=v_uphill/r_wheel
N_wheel=N_pedal*w_pedal/w_wheel


v_uphill=v_uphill*kph #单位转化

N_wheel,v_uphill
```

```python
%reset -f
%clear
from sympy import S,solve,zeros

w_ring,w_sun,w_arm=S('w_ring,w_sun,w_arm')

#基本公式
n_ring=80
n_sun=40
equ=w_ring*n_ring+w_sun*n_sun-(n_ring+n_sun)*w_arm

ans=zeros(6)

#case 1
w_arm_1=0
w_sun_1=40
equ1=equ.subs(w_arm,w_arm_1).subs(w_sun,w_sun_1)
ans[0]=solve(equ1)

#case 2
w_ring_2=50
w_sun_2=0
equ2=equ.subs(w_sun,w_sun_2).subs(w_ring,w_ring_2)
ans[1]=solve(equ2)

#case 3
w_ring_3=0
w_arm_3=50
equ3=equ.subs(w_arm,w_arm_3).subs(w_ring,w_ring_3)
ans[2]=solve(equ3)

#case 4
w_arm_4=-50
w_sun_4=40
equ4=equ.subs(w_arm,w_arm_4).subs(w_sun,w_sun_4)
ans[3]=solve(equ4)

#case 5
w_sun_5=-40
w_ring_5=50
equ5=equ.subs(w_sun,w_sun_5).subs(w_ring,w_ring_5)
ans[4]=solve(equ5)

#case 6
w_ring_6=-50
w_arm_6=50
equ6=equ.subs(w_arm,w_arm_6).subs(w_ring,w_ring_6)
ans[5]=solve(equ6)

for i in range(6):
    print(ans[i])
```
