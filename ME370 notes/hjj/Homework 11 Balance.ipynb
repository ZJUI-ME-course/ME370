### 三维单轴旋转平衡 (力矩平衡)

<div>

</div>
<div>

</div>

```python
m1=2*kg
m2=1*kg
m3=1.7*kg
theta1=120*du
theta2=45*du
theta3=250*du
r1=8*cm
r2=4*cm
r3=7*cm
l1=6*cm
l2=12*cm
l3=17*cm
lb=22*cm

ma=1*kg
mb=1*kg

w,rax,ray,rbx,rby=S('w,rax,ray,rbx,rby')
Ra=Matrix([rax,ray])
Rb=Matrix([rbx,rby])

R1=Matrix([r1*cos(theta1),r1*sin(theta1)])
R2=Matrix([r2*cos(theta2),r2*sin(theta2)])
R3=Matrix([r3*cos(theta3),r3*sin(theta3)])

#以A为旋转中
equ1=m1*R1*l1+m2*R2*l2+m3*R3*l3+mb*Rb*lb

#以B为旋转中心
equ2=m1*R1*(lb-l1)+m2*R2*(lb-l2)+m3*R3*(lb-l3)+mb*Ra*(lb-0)

#Ra.T #转置

solve(equ2),solve(equ1)
```

### 三维单轴旋转平衡  (力矩平衡)
<div>

</div>

```python
m1=1*kg
x1=-50*cm
y1=100*cm
z1=20*cm
m2=2*kg
x2=70*cm
y2=25*cm
z2=50*cm

mb=3*kg
zb=60*cm
rbx,rby=S('rbx,rby')

R1=Matrix([x1,y1])
R2=Matrix([x2,y2])
Rb=Matrix([rbx,rby])

equ=m1*R1*z1+m2*R2*z2+mb*Rb*zb
solve(equ)
```
