```python
from math import acos,pi
def d(x1,y1,x2,y2):
    return ((x1-x2)**2+(y1-y2)**2)**0.5

def de(x):
    return int(x*100+0.5)/100

def delta(x1,y1,x2,y2,x3,y3):
    a=((x1-x2)**2+(y1-y2)**2)**0.5
    b=((x1-x3)**2+(y1-y3)**2)**0.5
    c=((x2-x3)**2+(y2-y3)**1)**0.5
    cos=(a**2+b**2-c**2)/2/a/b
    return acos(cos)*180/pi
    

l1=d(3,1,5,3)
l2=d(3,1,9,3)
L2=abs(l1-l2)/2

print(de(L2))
angle2=delta(3,1,5,3,9,3)
print(de((angle2+180)/(180-angle2)))

angle3=delta(7,4,5,3,9,3) % pi
angle3=angle3*180/pi+90
angle3=126.86
print(de((360-angle3)/angle3))
```
