```python
from math import sin,cos,tan,asin,acos,atan,pi

rpm=2*pi/60
cm=0.01

a=50*cm
b=140*cm
w2=60*rpm
alpha2=0

def qz(x):
    if x>0:
        x=int(x*100+0.5)/100
    elif x<0:
        x=-int(abs(x)*100+0.5)/100   
    return x
#############################################
def d(a,b,w2,theta2,theta3):
    return a*cos(theta2)+b*cos(theta3)

def angle(a,b,theta2):
    theta3=asin(-a/b*sin(theta2))
    if theta3<0:
        theta3=2*pi+theta3
    return theta3
#############################################
def w3(a,b,w2,theta2,theta3):
    return -a*cos(theta2)/b/cos(theta3)*w2

def v(a,b,w2,theta2,theta3):
    return a*w2*(sin(theta3-theta2)/cos(theta3))
###############################################
def alpha(a,b,w2,w3,alpha2,theta2,theta3):
    alpha3=(-a/b*(alpha2*sin(theta2)-w2**2*sin(theta2))+w3**2*sin(theta3))/cos(theta3)
    return alpha3

def aa(a,b,w2,w3,alpha2,theta2,theta3):
    alpha3=alpha(a,b,w2,w3,alpha2,theta2,theta3)
    return -a*(alpha2*sin(theta2)+w2**2*cos(theta2))-b*(alpha3*sin(theta3)+w3**2*cos(theta3))
###############################################


for i in [60,90,180,240]:
    theta2=i*pi/180
    
    theta3=angle(a,b,theta2)
    distance=d(a,b,w2,theta2,theta3)
    
    omega3=w3(a,b,w2,theta2,theta3)
    velocity=v(a,b,w2,theta2,theta3)
    
    alpha3=alpha(a,b,w2,omega3,alpha2,theta2,theta3)
    acceleration=aa(a,b,w2,omega3,alpha2,theta2,theta3)

    
    print('theta3='+str(qz(theta3*180/pi)),'d='+str(qz(distance)))
    print()
    print('omega3='+str(qz(omega3)),'velocity='+str(qz(velocity)))
    print()
    print('alpha3='+str(qz(alpha3)),'acceleration='+str(qz(acceleration)))
    print()
    print()
```
